```python
import requests
headers = {"X-Api-Key": ""}

base = "https://api.sportsgameodds.com/v2/leagues"
```

```python
response = requests.get(base, headers=headers)
```

```python
response
```

```python
data = response.json()['data']
```

```python
len(data)
```

```python
LEAGUES = []
for league in data:
    LEAGUES.append(league['leagueID'])
```

```python
LEAGUES
```
