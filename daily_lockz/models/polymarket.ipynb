```python
import base64
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import serialization, hashes
from datetime import date, datetime as dt, timedelta
from dotenv import load_dotenv
import numpy as np
import os
import pandas as pd
import pytz
import requests
import warnings

warnings.filterwarnings("ignore")

load_dotenv()
KALSHI_API_KEY_ID = os.getenv("KALSHI_API_KEY")
PRIVATE_KEY_PATH = os.getenv("KALSHI_KEY_PATH_TEST")
# SPORTS_API_KEY = os.getenv("SPORTS_API_KEY")

DUPLICATE_PATH = "/Users/trevor/trevorscholz1/daily_lockz/models/test_placed.csv"
EDGE = 0.02

# get active sports
ACTIVE_SPORTS = "basketball_nba"
GAME_ODDS = GAME_ODDS
KALSHI_ODDS = KALSHI_ODDS


def convert_odds(prob):
    if prob >= 0.5:
        odds = -100 * prob / (1 - prob)
    else:
        odds = 100 * (1 - prob) / prob
    return round(odds)


def convert_time(utc):
    utc_time = dt.strptime(utc, "%Y-%m-%dT%H:%M:%SZ")

    utc_zone = pytz.timezone("UTC")
    utc_time = utc_zone.localize(utc_time)

    est_zone = pytz.timezone("US/Eastern")
    est_time = utc_time.astimezone(est_zone)

    now_est = dt.now(est_zone)
    is_today = est_time.date() == now_est.date()

    return est_time, is_today


def remove_juice(outcomes):
    outcomes_df = pd.DataFrame(outcomes)
    probs = 1 / outcomes_df["price"]
    fair_probs = probs / probs.sum()
    return outcomes_df.assign(fair_prob=fair_probs, raw=probs)


def game_odds(game, is_kalshi):
    game_odds = pd.DataFrame(
        columns=[
            "id",
            "sport",
            "h_team",
            "a_team",
            "book",
            "h_ml",
            "a_ml",
            "d_ml",
            "h_spread",
            "a_spread",
            "spreads_pt",
            "over",
            "under",
            "totals_pt",
            "time",
        ]
    )
    index = 0

    h_team = game["home_team"]
    a_team = game["away_team"]

    for book in game["bookmakers"]:
        game_odds.at[index, "id"] = game["id"] if not is_kalshi else game["sid"]
        game_odds.at[index, "sport"] = game["sport_key"]
        game_odds.at[index, "h_team"] = h_team
        game_odds.at[index, "a_team"] = a_team
        game_odds.at[index, "time"] = game["commence_time"]

        game_odds.at[index, "book"] = book["key"]
        for market in book["markets"]:
            if market["key"] == "h2h":
                outcomes = remove_juice(market["outcomes"])
                home_row = outcomes[outcomes["name"] == h_team]
                away_row = outcomes[outcomes["name"] == a_team]
                draw_row = outcomes[outcomes["name"] == "Draw"]
                game_odds.at[index, "h_ml"] = home_row["fair_prob"].item()
                game_odds.at[index, "a_ml"] = away_row["fair_prob"].item()
                game_odds.at[index, "d_ml"] = (
                    draw_row["fair_prob"].item() if not draw_row.empty else np.nan
                )

                game_odds.at[index, "h_ml_raw"] = home_row["raw"].item()
                game_odds.at[index, "a_ml_raw"] = away_row["raw"].item()
                game_odds.at[index, "d_ml_raw"] = (
                    draw_row["price"].item() if not draw_row.empty else np.nan
                )
            if market["key"] == "spreads":
                outcomes = remove_juice(market["outcomes"])
                home_row = outcomes[outcomes["name"] == h_team]
                away_row = outcomes[outcomes["name"] == a_team]
                game_odds.at[index, "h_spread"] = home_row["fair_prob"].item()
                game_odds.at[index, "a_spread"] = away_row["fair_prob"].item()
                game_odds.at[index, "spreads_pt"] = home_row["point"].item()

                game_odds.at[index, "h_spread_raw"] = home_row["raw"].item()
                game_odds.at[index, "a_spread_raw"] = away_row["raw"].item()
            if market["key"] == "totals":
                outcomes = remove_juice(market["outcomes"])
                over_row = outcomes[outcomes["name"] == "Over"]
                under_row = outcomes[outcomes["name"] == "Under"]
                game_odds.at[index, "over"] = over_row["fair_prob"].item()
                game_odds.at[index, "under"] = under_row["fair_prob"].item()
                game_odds.at[index, "totals_pt"] = over_row["point"].item()

                game_odds.at[index, "over_raw"] = over_row["raw"].item()
                game_odds.at[index, "under_raw"] = under_row["raw"].item()
        index += 1
    return game_odds


def kalshi_odds(game_id):
    found = False
    kalshi = None
    for game in KALSHI_ODDS:
        if game["id"] == game_id:
            found = True
            kalshi = game
            break
    if not found:
        return kalshi
    else:
        return game_odds(game, True)


def is_duplicate(sport, team, type, time):
    placed = pd.read_csv(DUPLICATE_PATH)
    placed["date_placed"] = pd.to_datetime(
        placed["date_placed"], errors="coerce"
    ).dt.date

    mask = (
        (placed["sport"] == sport)
        & (placed["team"] == team)
        & (placed["type"] == type)
        & (placed["time"] == time)
    )

    return mask.any()


def save_bet(sport, team, type, time):
    data = {
        "sport": [sport],
        "team": [team],
        "type": [type],
        "time": [time],
        "date_placed": [dt.date],
    }
    placed = pd.DataFrame(data=data)
    placed.to_csv(DUPLICATE_PATH, mode="a", index=False)


def clean_duplicates():
    cutoff = date.today() - timedelta(days=7)

    placed = pd.read_csv(DUPLICATE_PATH)
    placed["date_placed"] = pd.to_datetime(
        placed["date_placed"], errors="coerce"
    ).dt.date
    placed = placed[placed["date_placed"] >= cutoff]
    placed.to_csv(DUPLICATE_PATH, index=False)


def fliff_place_bet(sport, team, type, point, odds, time):
    if not is_duplicate(sport, team, type, time):
        print(sport, team, type, point, odds, time)
        save_bet(sport, team, type, time)
    else:
        print("SKIPPED DUPLICATE")


def kalshi_place_order(kalshi, raw_price):
    print(raw_price)

    def load_private_key(key_path):
        with open(key_path, "rb") as f:
            return serialization.load_pem_private_key(
                f.read(), password=None, backend=default_backend()
            )

    def create_signature(private_key, timestamp, method, path):
        path_without_query = path.split("?")[0]
        message = f"{timestamp}{method}{path_without_query}".encode("utf-8")
        signature = private_key.sign(
            message,
            padding.PSS(
                mgf=padding.MGF1(hashes.SHA256()), salt_length=padding.PSS.DIGEST_LENGTH
            ),
            hashes.SHA256(),
        )
        return base64.b64encode(signature).decode("utf-8")

    def post(
        private_key,
        api_key_id,
        path,
        payload,
        base_url="https://api.elections.kalshi.com",
    ):
        timestamp = str(int(dt.datetime.now().timestamp() * 1000))
        signature = create_signature(private_key, timestamp, "POST", path)

        headers = {
            "KALSHI-ACCESS-KEY": api_key_id,
            "KALSHI-ACCESS-SIGNATURE": signature,
            "KALSHI-ACCESS-TIMESTAMP": timestamp,
            "Content-Type": "application/json",
        }

        return requests.post(base_url + path, json=payload, headers=headers)

    if not is_duplicate(
        kalshi["sport"].item(),
        kalshi["team"].item(),
        kalshi["type"].item(),
        kalshi["time"].item(),
    ):
        payload = {
            "ticker": kalshi["ticker"],
            "side": "yes",
            "action": "buy",
            "count": 1,
            "yes_price": raw_price,
            "buy_max_cost": (raw_price * 8),
        }

        private_key = load_private_key(PRIVATE_KEY_PATH)
        response = post(
            private_key,
            KALSHI_API_KEY_ID,
            "/trade-api/v2/portfolio/orders",
            payload=payload,
        )
        print(response.json())
        save_bet(
            kalshi["sport"].item(),
            kalshi["team"].item(),
            kalshi["type"].item(),
            kalshi["time"].item(),
        )
    else:
        print("SKIPPED DUPLICATE")


def find_value(game_odds):
    fliff = game_odds[game_odds["book"] == "draftkings"]

    sport = game_odds["sport"][0]
    h_team = game_odds["h_team"][0]
    a_team = game_odds["a_team"][0]
    time = game_odds["time"][0]

    h_ml_median = game_odds["h_ml"].median()
    a_ml_median = game_odds["a_ml"].median()
    d_ml_median = game_odds["d_ml"].median()

    h_spread_median = game_odds[game_odds["spreads_pt"] == fliff["spreads_pt"].item()][
        "h_spread"
    ].median()
    a_spread_median = game_odds[game_odds["spreads_pt"] == fliff["spreads_pt"].item()][
        "a_spread"
    ].median()

    over_median = game_odds[game_odds["totals_pt"] == fliff["totals_pt"].item()][
        "over"
    ].median()
    under_median = game_odds[game_odds["totals_pt"] == fliff["totals_pt"].item()][
        "under"
    ].median()

    if fliff["h_ml"].item() < (h_ml_median - EDGE):
        fliff_place_bet(
            sport, h_team, "ML", 0, convert_odds(fliff["h_ml_raw"].item()), time
        )
    if fliff["a_ml"].item() < (a_ml_median - EDGE):
        fliff_place_bet(
            sport, a_team, "ML", 0, convert_odds(fliff["a_ml_raw"].item()), time
        )
    if fliff["d_ml"].item() < (d_ml_median - EDGE):
        fliff_place_bet(
            sport, "Draw", "ML", 0, convert_odds(fliff["d_ml_raw"].item()), time
        )

    if fliff["h_spread"].item() < (h_spread_median - EDGE):
        fliff_place_bet(
            sport, h_team, "Spread", 0, convert_odds(fliff["h_spread_raw"].item()), time
        )
    if fliff["a_spread"].item() < (a_spread_median - EDGE):
        fliff_place_bet(
            sport, a_team, "Spread", 0, convert_odds(fliff["a_spread_raw"].item()), time
        )

    if fliff["over"].item() < (over_median - EDGE):
        fliff_place_bet(
            sport, h_team, "Over", 0, convert_odds(fliff["over_raw"].item()), time
        )
    if fliff["under"].item() < (under_median - EDGE):
        fliff_place_bet(
            sport, h_team, "Under", 0, convert_odds(fliff["under_raw"].item()), time
        )

    kalshi = kalshi_odds(game_odds["id"][0])
    display(kalshi)
    if kalshi is not None:
        if kalshi["h_ml"].item() < h_ml_median:
            kalshi_place_order(kalshi, kalshi["h_ml_raw"].item())
        if kalshi["a_ml"].item() < a_ml_median:
            kalshi_place_order(kalshi, kalshi["a_ml_raw"].item())
        if kalshi["d_ml"].item(kalshi) < d_ml_median:
            kalshi_place_order(kalshi), kalshi["d_ml_raw"].item()
```
