```python
import datetime as dt
import numpy as np
import pandas as pd
import pytz
import requests

BETS = pd.DataFrame(columns=["sport", "team", "type", "point", "odds", "time"])
EDGE = 0.02


def convert_time(utc):
    utc_time = dt.strptime(utc, "%Y-%m-%dT%H:%M:%SZ")

    utc_zone = pytz.timezone("UTC")
    utc_time = utc_zone.localize(utc_time)

    est_zone = pytz.timezone("US/Eastern")
    est_time = utc_time.astimezone(est_zone)

    now_est = dt.now(est_zone)
    is_today = est_time.date() == now_est.date()

    return est_time, is_today


# get active sports

# get sport odds

# loop through games


# create table of game odds without juice, and the raw odds
def remove_juice(outcomes):
    outcomes_df = pd.DataFrame(outcomes)
    probs = 1 / outcomes_df["price"]
    fair_probs = probs / probs.sum()
    return outcomes_df.assign(fair_prob=fair_probs)


def game_odds(game):
    game_odds = pd.DataFrame(
        columns=[
            "sport",
            "h_team",
            "a_team",
            "book",
            "h_ml",
            "a_ml",
            "d_ml",
            "h_spread",
            "a_spread",
            "spreads_pt",
            "over",
            "under",
            "totals_pt",
        ]
    )
    index = 0
    sport = game["sport_key"]
    home_team = game["home_team"]
    away_team = game["away_team"]

    for book in game["bookmakers"]:
        game_odds.at[index, "book"] = book["key"]
        for market in book["markets"]:
            if market["key"] == "h2h":
                outcomes = remove_juice(market["outcomes"])
                home_row = outcomes[outcomes["name"] == home_team]
                away_row = outcomes[outcomes["name"] == away_team]
                draw_row = outcomes[outcomes["name"] == "Draw"]
                game_odds.at[index, "h_ml"] = home_row["fair_prob"].item()
                game_odds.at[index, "a_ml"] = away_row["fair_prob"].item()
                game_odds.at[index, "d_ml"] = (
                    draw_row["fair_prob"].item() if not draw_row.empty else np.nan
                )
            if market["key"] == "spreads":
                outcomes = remove_juice(market["outcomes"])
                home_row = outcomes[outcomes["name"] == home_team]
                away_row = outcomes[outcomes["name"] == away_team]
                game_odds.at[index, "h_spread"] = home_row["fair_prob"].item()
                game_odds.at[index, "a_spread"] = away_row["fair_prob"].item()
                game_odds.at[index, "spreads_pt"] = home_row["point"].item()
            if market["key"] == "totals":
                outcomes = remove_juice(market["outcomes"])
                over_row = outcomes[outcomes["name"] == "Over"]
                under_row = outcomes[outcomes["name"] == "Under"]
                game_odds.at[index, "over"] = over_row["fair_prob"].item()
                game_odds.at[index, "under"] = under_row["fair_prob"].item()
                game_odds.at[index, "totals_pt"] = over_row["point"].item()
        index += 1
    return game_odds


# find fliff odds
def fliff_value(game_odds):
    fliff = game_odds[game_odds["book"] == "draftkings"]

    if fliff["h_ml"].item() < game_odds["h_ml"].median():
        print("Bet Found", fliff["h_ml"])
    if fliff["a_ml"].item() < game_odds["a_ml"].median():
        print("Bet Found", fliff["a_ml"])
    if fliff["d_ml"].item() < game_odds["d_ml"].median():
        print("Bet Found", fliff["d_ml"])

    spreads_line = game_odds[game_odds["spreads_pt"] == fliff["spreads_pt"].item()]
    if fliff["h_spread"].item() < spreads_line["h_spread"].median():
        print("Bet Found", fliff["h_spread"])
    if fliff["a_spread"].item() < spreads_line["a_spread"].median():
        print("Bet Found", fliff["a_spread"])

    totals_line = game_odds[game_odds["totals_pt"] == fliff["totals_pt"].item()]
    if fliff["over"].item() < totals_line["over"].median():
        print("Bet Found", fliff["over"])
    if fliff["under"].item() < totals_line["under"].median():
        print("Bet Found", fliff["under"])


def polymarket_find_game(game):
    query = (
        game["sport_title"] + "%" + game["home_team"] + "%" + game["away_team"]
    ).replace(" ", "_")
    response = requests.get(f"https://gamma-api.polymarket.com/public-search?q={query}&search_profiles=false")
    data = response.json()
    
    for event in data["events"]:
        if event["startTime"] == game["commence_time"]:
            return event
    return None


def polymarket_value(game_odds, event):
    for market in event["markets"]:
        if market["slug"].split('-')[1] == market["slug"].split('-')[-1]:
            if market < game_odds["h_ml"].median():


# if so, save
```
